```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import variation
```

```python
Accra=pd.read_csv("accra.csv")
Accra.replace(to_replace = -999, value =np.nan, inplace = True)
Ada=pd.read_csv("ada.csv")
Ada.replace(to_replace = -999, value =np.nan, inplace = True)

Axim=pd.read_csv("axim.csv")
Axim.replace(to_replace = -999, value =np.nan, inplace = True)
Saltpond=pd.read_csv("saltpond.csv")
Saltpond.replace(to_replace = -999, value =np.nan, inplace = True)
Takoradi=pd.read_csv("takoradi.csv")
Takoradi.replace(to_replace = -999, value =np.nan, inplace = True)

Koforidua=pd.read_csv("koforidua.csv")
Koforidua.replace(to_replace = -999, value =np.nan, inplace = True)
Kumasi=pd.read_csv("kumasi.csv")
Kumasi.replace(to_replace = -999, value =np.nan, inplace = True)
Bekwai=pd.read_csv("sefwi_bekwai.csv")
Bekwai.replace(to_replace = -999, value =np.nan, inplace = True)
Akuse=pd.read_csv("akuse.csv")
Akuse.replace(to_replace = -999, value =np.nan, inplace = True)

Sunyani=pd.read_csv("sunyani.csv")
Sunyani.replace(to_replace = -999, value =np.nan, inplace = True)

KeteKrachi=pd.read_csv("ketekrachi.csv")
KeteKrachi.replace(to_replace = -999, value =np.nan, inplace = True)
Wenchi=pd.read_csv("wenchi.csv")
Wenchi.replace(to_replace = -999, value =np.nan, inplace = True)
               
Otherrr=pd.read_csv("totalrr.csv")               
Otherrainy=pd.read_csv("rainydays.csv") 
```

```python
from sklearn.metrics import r2_score

fig,axs = plt.subplots(3,1, figsize =(20,25))

axs[0].plot(Saltpond.YEARS,Saltpond.Minor_amount, ms=10, mec="k", lw=3)
z = np.polyfit(Saltpond.YEARS,Saltpond.Minor_amount, 1)
y_hat = np.poly1d(z)(Saltpond.YEARS)
axs[0].plot(Saltpond.YEARS, y_hat, "r-", lw=3)
axs[0].set_title("(a)   ",x=0.03,y=0.93,fontsize="25")
axs[0].set_xlabel("Years ",fontsize="25")
axs[0].set_ylabel("Rainfall Amount (mm)",fontsize="25")
text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(Saltpond.Minor_amount,y_hat):0.3f}$"
#plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     #fontsize=14, verticalalignment='top')
axs[0].text(0.05, 0.95, text, transform=axs[0].transAxes,
     fontsize=25, verticalalignment='top')
   

axs[1].plot(Takoradi.YEARS,Takoradi.Minor_amount, ms=10, mec="k", lw=3)
p = np.polyfit(Takoradi.YEARS,Takoradi.Minor_amount, 1)
p_hat = np.poly1d(p)(Takoradi.YEARS)


axs[1].set_title("(b)   ",x=0.03,y=0.93,fontsize="25")
axs[1].plot(Takoradi.YEARS, p_hat, "r-", lw=3)
text = f"$y={p[0]:0.3f}\;x{p[1]:+0.3f}$\n$R^2 = {r2_score(Takoradi.Minor_amount,p_hat):0.3f}$"
#plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     #fontsize=14, verticalalignment='top')
axs[1].text(0.05, 0.95, text, transform=axs[1].transAxes,
     fontsize=14, verticalalignment='top')   
axs[1].set_xlabel("Years ",fontsize="25")
axs[1].set_ylabel("Rainfall Amount (mm)",fontsize="25")


axs[2].plot(Axim.YEARS,Axim.Minor_amount, ms=10, mec="k", lw=3)
k = np.polyfit(Axim.YEARS,Axim.Minor_amount, 1)
m_hat = np.poly1d(k)(Axim.YEARS)


axs[2].set_title("(c)   ",x=0.03,y=0.93,fontsize="25")
axs[2].plot(Axim.YEARS, m_hat, "r-", lw=3)
text = f"$y={k[0]:0.3f}\;x{k[1]:+0.3f}$\n$R^2 = {r2_score(Axim.Minor_amount,m_hat):0.3f}$"
#plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     #fontsize=14, verticalalignment='top')
axs[2].text(0.05, 0.95, text, transform=axs[2].transAxes,
     fontsize=25, verticalalignment='top')
   
axs[2].set_xlabel("Years ",fontsize="25")
axs[2].set_ylabel("Rainfall Amount (mm)",fontsize="25")


fig.suptitle("Figure 2: Minor Seasonal Rainfall over the  Western Coast of Ghana\n(a) Saltpond (b) Takoradi (c) Axim ",y=0.070,fontsize="25")
for ax in axs.flat:
    plt.setp(ax.get_xticklabels(),fontsize="25", rotation=15)
    
for ax in axs.flat:
    plt.setp(ax.get_yticklabels(),fontsize="25")    
    
    
    
for ax in axs.flat:
    ax.legend()
        
    
plt.savefig("WesternCoastRain.png")
```
